```python
import pandas as pd
df = pd.read_excel('2024 Crosswalk-Engineering summer camps application form fields.xlsx',
                  sheet_name=None)

camp_info_df = df["Pref"]
camp_info_df.head()
```

## Truncate Excel sheet to show camp, start and end date

```python
filter = ['value options', 'start date', 'end date']
camp_date_df = camp_info_df[filter]
camp_date_df = camp_date_df[camp_info_df['value options'].notnull()]
camp_date_df
```

# Read Master Excel Sheet in

```python
camper_data_df = pd.read_excel("2024_campers.xlsx")
```

## Filter excel sheet for registered campers in each camp

```python
# Filter entries by those who have been fully registered
registered_camper_filter = camper_data_df['assigned_session'].notnull() 
camper_data_df = camper_data_df[registered_camper_filter]

# Filter dataframe with just a few columns of interest
roster_filters = ["assigned_session",'camper_fn', 'camper_ln', 'camper_pref_name',
                  'camper_gender', 'camper_pronouns', 'camper_year_in_school',
                  'camper_shirt_size', 'waitlist_notification_sent', 'will_commute', 
                  'roommate_request', 'camper_room_single', 'photo_release',
                  'medical_allergies', 'medical_allergies_anaphylaxis', 'medical_consent', 'additional_info']
camper_data_df[roster_filters].head()
```

```python
import re
def multiple_replace(string, rep_dict):
    pattern = re.compile("|".join([re.escape(k) for k in sorted(rep_dict,key=len,reverse=True)]), flags=re.DOTALL)
    return pattern.sub(lambda x: rep_dict[x.group(0)], string)
    
# Get unique camp names
original_camp_names = camper_data_df['assigned_session'].unique()
camps = original_camp_names.copy()

# Specify chars to replace
chars2replace = {c : '' for c in [ ':', ',', '[', ']', '*', '-']}

# For each camp, replace the specified chars, and truncate the first 4 chars of each substring and the total string to 22 
camps = [''.join([c[:4] 
                  for c in multiple_replace(c, chars2replace).split(' ')])[:22] 
                         for c in camps]
# Print truncated names to see
print("Truncated camp names")
camps
```

```python
# Group campers by camps
camp_dfs = [camper_data_df.loc[ camper_data_df['assigned_session'] == c][roster_filters] for c in original_camp_names]

# Add Camp date sheet
camps = ['camp_dates'] + camps
camp_dfs = [camp_date_df] + camp_dfs

# Helper function to adjust column lenths
def write_sheet_and_adjust_cols(df, writer, sheet_name):

    df.to_excel(writer, sheet_name=sheet_name, index=False) 

    # get the XlsxWriter workbook and worksheet objects
    workbook = writer.book
    worksheet = writer.sheets[sheet_name]
    
    # adjust the column widths based on the content
    for i, col in enumerate(df.columns):
        width = max(df[col].apply(lambda x: len(str(x))).max(), len(col))
        worksheet.set_column(i, i, width)
    
output_xlsx_path = "campers_filtered_by_camp.xlsx"
with pd.ExcelWriter(output_xlsx_path, mode="w", engine="xlsxwriter") as writer:
    # [cur_df.to_excel(writer, sheet_name=f"{name}_roster")  for name, cur_df in zip(camps, camp_dfs)]
    [write_sheet_and_adjust_cols(cur_df, writer, sheet_name=f"{name}_roster")  for name, cur_df in zip(camps, camp_dfs)]
    
```
